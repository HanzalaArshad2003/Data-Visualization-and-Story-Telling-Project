```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("euro-daily-hist_1999_2020.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.rename(columns={'Period\\Unit:':'Time'},inplace=True)#Backslash two times BCZ of some reason
```

```python
dataset.rename(columns={'[US dollar ]':'USD'},inplace=True)
```

```python
dataset.info()
```

```python
dataset.Time=pd.to_datetime(dataset.Time)
```

```python
dataset.sort_values("Time",inplace=True)
```

```python
dataset.reset_index(inplace=True,drop=True)
```

```python
dataset.head()
```

```python
euro_to_dollar=dataset[["Time","USD"]].copy()
```

```python
euro_to_dollar
```

```python
euro_to_dollar["USD"].value_counts()
```

```python
euro_to_dollar=euro_to_dollar[euro_to_dollar.USD != "-"]#To print without hyphen
```

```python
euro_to_dollar["USD"].value_counts()
```

```python
euro_to_dollar['USD'] = euro_to_dollar['USD'].astype(float)
```

```python
euro_to_dollar.info()
```

```python
euro_to_dollar.head()
```

```python
plt.plot(euro_to_dollar["Time"],euro_to_dollar["USD"])
plt.show()
```

```python
euro_to_dollar["USD"].rolling(7).mean().head(8)#& rolling means 1st 6 vlues mean 
```

```python
euro_to_dollar["rolling_mean"]=euro_to_dollar["USD"].rolling(30).mean()#30 wala liya kyunke osmain details shi se bta chl rhi hein
```

```python
euro_to_dollar.head(40)
```

```python
Financial_Crisis=euro_to_dollar[(euro_to_dollar.Time.dt.year >= 2006) & (euro_to_dollar.Time.dt.year <= 2009)]
Financial_Crisis_7_8=euro_to_dollar[(euro_to_dollar.Time.dt.year >= 2007) & (euro_to_dollar.Time.dt.year <= 2008)]
```

```python
import matplotlib.style as style
style.use("fivethirtyeight")
```

```python
fig, ax=plt.subplots(figsize=(8,3))

ax.plot(Financial_Crisis.Time,Financial_Crisis["rolling_mean"],linewidth=1.5,color="green")
ax.plot(Financial_Crisis_7_8.Time,Financial_Crisis_7_8["rolling_mean"],linewidth=2,color="red")

plt.xticks(rotation=90)
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(8,3))

ax.plot(Financial_Crisis.Time,Financial_Crisis["rolling_mean"],linewidth=1.5,color="green")
ax.plot(Financial_Crisis_7_8.Time,Financial_Crisis_7_8["rolling_mean"],linewidth=2,color="red")

#plt.xticks(rotation=90)
ax.set_xticklabels(["2006","","2007","","2008","","2009","","2010"],alpha=0.3,fontdict={"fontsize":11})
ax.set_yticklabels(["","1.2","1.3","1.4","$ 1.5"],alpha=0.3,fontdict={"fontsize":11})

#For Text in anywhere in graph
ax.text(pd.to_datetime("2006-01-1"),1.7,"EURO-USD peaked at 1.59$ during 2008-2009",weight="bold")
ax.text(pd.to_datetime("2006-01-1"),1.65,"EURO-USD between 2006 to 2010")

ax.text(pd.to_datetime("2005-5-1"),1.07,"(C) Hanzala Arshad" + " "*72 + "Source: European Central Bank",backgroundcolor="#4d4d4d")
ax.axvspan(xmin=pd.to_datetime("2008-4-1"),xmax=pd.to_datetime("2008-9-1"),color="grey",alpha=0.4,ymin=0.08)
plt.show()
```
