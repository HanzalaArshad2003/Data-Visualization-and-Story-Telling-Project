```python
import pandas as pd
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("euro-daily-hist_1999_2020.csv")
```

```python
dataset.head()
```

```python
dataset.rename(columns={"Period\\Unit:":"Time"},inplace=True)
```

```python
dataset.rename(columns={"[US dollar ]":"USD"},inplace=True)
```

```python
dataset.head()
```

```python
#Now change Data Types
dataset.Time=pd.to_datetime(dataset.Time)
```

```python
dataset.sort_values("Time",inplace=True)
```

```python
dataset.reset_index(inplace=True,drop=True)
```

```python
dataset.head()
```

```python
euro_to_dollar=dataset[["Time","USD"]].copy()
```

```python
euro_to_dollar.head()
```

```python
euro_to_dollar["USD"].value_counts()                                                                   
```

```python
euro_to_dollar=euro_to_dollar[euro_to_dollar.USD != "-"]
```

```python
euro_to_dollar.head()
```

```python
euro_to_dollar["USD"]=euro_to_dollar["USD"].astype(float)
```

```python
euro_to_dollar.info()
```

```python
euro_to_dollar["rolling_mean"]=euro_to_dollar["USD"].rolling(30).mean()
```

```python
euro_to_dollar.head(40)
```

```python
financial_crisis_1_9=euro_to_dollar[(euro_to_dollar.Time.dt.year >= 2001) & (euro_to_dollar.Time.dt.year < 2009)]
financial_crisis_9_17=euro_to_dollar[(euro_to_dollar.Time.dt.year >= 2009) & (euro_to_dollar.Time.dt.year < 2017)]
financial_crisis_17_21=euro_to_dollar[(euro_to_dollar.Time.dt.year >= 2017) & (euro_to_dollar.Time.dt.year <= 2020)]
```

```python
financial_crisis_1_9
```

```python
import matplotlib.style as style
style.use("fivethirtyeight")
```

```python
fig, ax =plt.subplots(1,3,figsize=(12,3))

ax[0].plot(financial_crisis_1_9.Time,financial_crisis_1_9["rolling_mean"],linewidth=2,color="blue")
ax[0].set_xticklabels(["","2001","","2003","","2005","","2007","","2009"],alpha=0.3,fontdict={"fontsize":11})
ax[0].set_yticklabels(["","1.0","1.2","1.4","1.6 $"],alpha=0.3,fontdict={"fontsize":11})
ax[0].text(pd.to_datetime("1999-01-1"),1.7,"EURO-USD rate averaged 1.22 under the last three US presidents",weight="bold")
ax[0].text(pd.to_datetime("1999-01-1"),1.63,"EURO-USD exchange rate under George W.Bush (2001-2009),Barack Obama (2009-2017) and Donald Trump (2017-2021)")
ax[0].text(pd.to_datetime("2004"),0.68,"BUSH",weight="bold",color="blue")
ax[0].text(pd.to_datetime("2003"),0.6,"(2001-2009)",alpha=0.3)

ax[1].plot(financial_crisis_9_17.Time, financial_crisis_9_17["rolling_mean"], linewidth=2, color="green")
ax[1].set_xlim(pd.to_datetime("2009-01-01"), pd.to_datetime("2017-12-31"))
ax[1].set_ylim(1.0, 1.6)
ax[1].set_xticklabels(["", "2009", "", "2011", "", "2013", "", "2015", "", "2017"], alpha=0.3, fontdict={"fontsize":11})
ax[1].set_yticklabels(["1.0", "1.2", "1.4", "1.6 $"], alpha=0.3, fontdict={"fontsize":11})
ax[1].text(pd.to_datetime("2012"), 0.9, "OBAMA", weight="bold", color="blue")
ax[1].text(pd.to_datetime("2011-6-1"), 0.85, "(2009-2017)", alpha=0.3)

ax[2].plot(financial_crisis_17_21.Time, financial_crisis_17_21["rolling_mean"], linewidth=2, color="red")
ax[2].set_xlim(pd.to_datetime("2017-01-01"), pd.to_datetime("2021-12-31"))
ax[2].set_ylim(1.0, 1.6)
ax[2].set_xticklabels(["","","2017", "", "2018", "", "2019", "", "2020"], alpha=0.3, fontdict={"fontsize":11})
ax[2].set_yticklabels(["1.0", "1.2", "1.4", "1.6 $"], alpha=0.3, fontdict={"fontsize":11})
ax[2].text(pd.to_datetime("2018-6-1"), 0.9, "TRUMP", weight="bold", color="blue")
ax[2].text(pd.to_datetime("2018-3-15"), 0.85, "(2017-2021)", alpha=0.3)

fig, axs =plt.subplots(figsize=(12,3))
axs.plot(financial_crisis_1_9.Time,financial_crisis_1_9["rolling_mean"],linewidth=2,color="blue")
axs.plot(financial_crisis_9_17.Time,financial_crisis_9_17["rolling_mean"],linewidth=2,color="green")
axs.plot(financial_crisis_17_21.Time,financial_crisis_17_21["rolling_mean"],linewidth=2,color="red")
axs.set_xticklabels([])
axs.tick_params(bottom=0,left=0)
axs.set_yticklabels(["","1.0","1.2","1.4","1.6 $"],alpha=0.3,fontdict={"fontsize":11})
axs.text(pd.to_datetime("1999-1-1"),0.7,"(C) Hanzala Arshad" + " "*110 + "Source: European Central Bank",backgroundcolor="#4d4d4d")

plt.tight_layout()
plt.show()
```
